# 머신러닝과 딥러닝

* 머신러닝과 딥러닝
    * 심층 학습(Deep learning)  < 기계 학습 < 인공 지능
    * 학습이란 "한 컴퓨터 프로그램이 어떤 과제류 T에 속하는 과제들을 수행하며 그 수행의 성과를 측정한 측도가 P라고 할 때. 만일 어떤 경험 E 때문에 T의 어떤 과제에 대한 성과 측도 P가 개선되었다면, 그 컴퓨터 프로그램은 경험 E로부터 학습한다고 말할 수 있다" - mitchell, 1997

* 만능은 아니다!
    * 개리 마르커스 교수의 에세이(https://arxiv.org/ftp/arxiv/papers/1801/1801.00631.pdf)에서 몇 가지 발췌
    * 많은 데이터 필요(데이터는 신뢰할 수 있는 데이터인가?)
    * 재활용 힘듬
    * 충분히 투명하지 않음
    * 인과관계와 상관관계의 구분이 어려움
    * 신뢰성있는 엔지니어링의 어려움

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
주가 = [np.random.randint(10, 50) + i * 2 for i in range(100)]
```

```python
독립 = pd.DataFrame(np.arange(1, 101))
종속 = pd.DataFrame(주가)

종속
독립
```

```python
import tensorflow as tf # 구글이 2015년 오픈소스로 공개한 딥러닝 라이브러리
```

```python
X = tf.keras.layers.Input(shape=[1]) # 독립변수의 col
Y = tf.keras.layers.Dense(1)(X) # 종속변수의 col(뉴런의 개수)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse') # MSE(Mean squared error)
```

```python
model.fit(독립, 종속, epochs=10000, verbose=0)
model.fit(독립, 종속, epochs=10)
```

```python
model.predict([50]) # 예측

# 2 * x + 25
```

```python
model.predict(독립) # 예측
```

```python
종속
```

```python
model.get_weights()
```

```python
2.0389647 * 50 + 27.04446

# 2 * x + 25
```

```python
plt.plot(np.arange(1, 101), 주가)
plt.plot(np.arange(1, 101), 2.0389647 * np.arange(1, 101) + 27.04446)

plt.show()
```

```python
오차값 = 종속 - model.predict(독립)
오차값
```

```python
오차의제곱 = 오차값 ** 2
오차의제곱
```

```python
오차의제곱.sum() / 100
```

## 단순 데이터의 히든레이어

```python
유입수 = [2, 5, 10, 20, 40, 80]
광고액 = [1, 2, 4, 6, 8, 10]
순익 = [1, 1.5, 3, 10, 20, 60]
```

```python
plt.plot(np.arange(1, 7), 유입수, label='a')
plt.plot(np.arange(1, 7), 광고액, label='b')
plt.plot(np.arange(1, 7), 순익, label='c')

plt.legend() # 범례
plt.show()
```

```python
독립 = pd.DataFrame({
    '유입수': 유입수,
    '광고액': 광고액
})

종속 = pd.DataFrame({
    '순익': 순익
})

독립
종속
```

```python
X = tf.keras.layers.Input(shape=[2]) # 독립변수의 col
H = tf.keras.layers.Dense(2, activation='swish')(X)
# activation은 뉴런 출력에 적용할 함수(대표적으로 swish, sigmoid, softmax)
# H = tf.keras.layers.Dense(2, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H) # 종속변수의 col(뉴런의 개수)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse') # MSE(Mean squared error)
```

```python
model.fit(독립, 종속, epochs=10000, verbose=0)
model.fit(독립, 종속, epochs=10)
```

```python
독립
```

```python
종속
```

```python
model.get_weights()
```

```python
model.predict([[10, 4]])
```

```python
model.predict(독립)
```

```python
plt.plot(np.arange(1, 7), model.predict(독립), label='a')
plt.plot(np.arange(1, 7), 종속, label='b')

plt.legend()
plt.show()
```
